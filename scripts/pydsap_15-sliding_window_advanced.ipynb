## **ðŸ“ Snippet 1 â€” Longest Substring Without Repeating Characters**

*A classic variable-window pattern: expand â†’ shrink while duplicates exist.*

```python
def longest_unique_substring(s: str) -> int:
    left = 0
    seen = set()
    max_len = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
        max_len = max(max_len, right - left + 1)

    print("Longest Unique Substring Length:", max_len)

longest_unique_substring("abcabcbb")
# Output: 3 ("abc")
```

## **ðŸ“ Snippet 2 â€” Longest Substring with At Most K Distinct Characters**

*Window expands until >K distinct chars, then shrink until valid.*

```python
from collections import defaultdict

def longest_k_distinct(s: str, k: int) -> int:
    left = 0
    freq = defaultdict(int)
    max_len = 0

    for right in range(len(s)):
        freq[s[right]] += 1

        while len(freq) > k:
            freq[s[left]] -= 1
            if freq[s[left]] == 0:
                del freq[s[left]]
            left += 1

        max_len = max(max_len, right - left + 1)

    print(f"Longest Substring With â‰¤{k} Distinct Characters:", max_len)

longest_k_distinct("eceba", 2)
# Output: 3 ("ece")
```

## **ðŸ“ Snippet 3 â€” Minimum Window Substring**

*Hardest sliding-window classic â€” shrink window while keeping it valid.*

```python
from collections import Counter

def min_window(s: str, t: str) -> str:
    if not t or not s:
        return ""

    need = Counter(t)
    have = {}
    required = len(need)
    formed = 0

    left = 0
    ans = (float("inf"), 0, 0)

    for right, ch in enumerate(s):
        have[ch] = have.get(ch, 0) + 1

        if ch in need and have[ch] == need[ch]:
            formed += 1

        while left <= right and formed == required:
            if right - left + 1 < ans[0]:
                ans = (right - left + 1, left, right)

            have[s[left]] -= 1
            if s[left] in need and have[s[left]] < need[s[left]]:
                formed -= 1

            left += 1

    if ans[0] == float("inf"):
        return ""

    result = s[ans[1]:ans[2] + 1]
    print("Minimum Window Substring:", result)
    return result

min_window("ADOBECODEBANC", "ABC")
# Output: "BANC"
```

## **ðŸ“ Snippet 4 â€” Sliding Window Maximum (Using Deque)**

*Maintain a deque of useful elements â†’ O(n) solution.*

```python
from collections import deque

def sliding_window_max(nums, k):
    dq = deque()
    result = []

    for i in range(len(nums)):
        # remove out-of-window elements
        if dq and dq[0] == i - k:
            dq.popleft()

        # maintain decreasing order in deque
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        dq.append(i)

        # first valid window ends at index k-1
        if i >= k - 1:
            result.append(nums[dq[0]])

    print("Sliding Window Maximum:", result)

sliding_window_max([1,3,-1,-3,5,3,6,7], 3)
# Output: [3,3,5,5,6,7]
```

## **âœ… Takeaways**

- Advanced Sliding Window is used when:
    - Window **expands and shrinks dynamically**
    - Conditions depend on **frequency**, **distinct count**, or **match completeness**
- Typical structure:
    - Expand right pointer
    - Update condition
    - Shrink left pointer until valid
- Solves many problems:
    - Longest unique substring
    - K distinct characters
    - Minimum window substring
    - Sliding window max/min
- Time complexity usually **O(n)** or **O(n log n)** depending on data structure
